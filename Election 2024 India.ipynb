```python
#Loading the ELECTION dataset 2024
import pandas as pd
import matplotlib.pyplot as plt
import chardet

# Detect encoding
with open('C:/indian election dataset/eci_data_2024.csv', 'rb') as file:
    raw_data = file.read()
    result = chardet.detect(raw_data)
    encoding = result['encoding']

print(f"Detected encoding: {encoding}")
data = pd.read_csv('C:/indian election dataset/eci_data_2024.csv', encoding='unicode_escape')
```

```python
#Checking for null items in the dataset
missing_rows = data[data.isnull().any(axis=1)]
missing_rows
```

```python
missing_values = data.isnull().sum()
missing_values
```

```python
data['State'] = data['State'].astype('category')
data['Party'] = data['Party'].astype('category')
```

```python
import numpy as np
data['Total Votes'] = pd.to_numeric(data['Total Votes'], errors='coerce')
```

```python
# visualization
import warnings
warnings.filterwarnings("ignore", category=UserWarning)
import seaborn as sns
plt.rcParams['font.family'] = 'Arial'
plt.figure(figsize=(10, 8))
# Creating a bar plot for vote distribution
sns.barplot(x='State', y='Total Votes', hue='Party', data=data)
plt.title('Vote Distribution by State and Party')
plt.xticks(rotation=90)
plt.show()
```
